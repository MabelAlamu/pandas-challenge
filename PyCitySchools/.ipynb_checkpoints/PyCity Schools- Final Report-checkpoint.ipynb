```python
#Importing dependency
import pandas as pd

#Reading in both csv files
school_path = "schools_complete.csv"
student_path = "students_complete.csv"

school_data = pd.read_csv(school_path)
student_data = pd.read_csv(student_path)

#Merging both datasets on school name
school_df = pd.merge(student_data, school_data, how="left", on=["school_name","school_name"])
school_df
```

```python
#Checking for null values
school_df.isnull().sum()
```

```python
#Checking for duplicated rows
school_df.duplicated().any()
```

```python
#Checking data type of each column
school_df.dtypes
```

```python
#Checking for inconsistent values
school_df['gender'].value_counts(dropna=False)
```

```python
#Checking for inconsistent values
school_df['grade'].value_counts(dropna=False)
```

```python
#Checking for inconsistent values and inconsistent usage of upper and lower cases  
school_df['school_name'].value_counts(dropna=False)
```

```python
#Checking for inconsistent values and inconsistent usage of upper and lower cases 
school_df['type'].value_counts(dropna=False)
```

# District Summary

```python
#Calculating the total number of schools
total_schools = school_df["school_name"].nunique()
total_schools
```

```python
#Calculating the total number of students
total_students = school_df["student_name"].count()
total_students
```

```python
#Calculating the average math score in all schools
avg_math = round(school_df["math_score"].mean(),2)
avg_math
```

```python
#Calculating the average reading score in all schools
avg_read = round(school_df["reading_score"].mean(),2)
avg_read
```

```python
#Calculating the percentage of students that scored 70 and above in math
passed_math = round(((school_df[school_df.math_score >= 70]["student_name"].count())/total_students)*100,2)
passed_math
```

```python
#Calculating the percentage of students that scored 70 and above in reading
passed_reading = round(((school_df[school_df.reading_score >= 70]["student_name"].count())/total_students)*100,2)
passed_reading
```

```python
#Calculating the percentage of students who passed math and reading (% Overall Passing)
overall = []
for i in school_df.index:
    math_70 = school_df['math_score'][i]
    read_70 = school_df['reading_score'][i]
    if (math_70 >= 70) & (read_70 >= 70):
        overall.append(school_df['student_name'])

overall_passed= round((len(overall)/total_students)*100,2)
overall_passed
```

```python
#Creating a new table for the summary statistics
district_summary= pd.DataFrame({'Total Schools':  [total_schools],
                  'Total Students': [total_students],
                 'Average Math Score':[avg_math],
                 'Average Reading Score': [avg_read],
                              '% Passing Math':[passed_math],
                              '% Passing Reading':[passed_reading], 
                              '% Overall Passing':[overall_passed]})
district_summary
```

# School Summary

```python
unique_school= school_df.groupby(['school_name'])
unique_school
```

```python
#Identifying 
school_type= unique_school["type"].first()
school_type
```

```python
total_students= unique_school["size"].first()
total_students
```

```python
total_budget = unique_school["budget"].first()
total_budget
```

```python
student_budget = total_budget/total_students
student_budget
```

```python
avg_readscore = round(unique_school["reading_score"].mean(),2)
avg_readscore
```

```python
avg_mathscore = round(unique_school["math_score"].mean(),2)
avg_mathscore
```

```python
# Calculate the percentange of passing math score for each school
pass_math = round((school_df[school_df.math_score >= 70].groupby(["school_name"]).size()/total_students)*100,2)
pass_math
```

```python
# Calculate the percentange of passing reading score for each school
pass_reading = round((school_df[school_df.reading_score >= 70].groupby(["school_name"]).size()/total_students)*100,2)
pass_reading
```

```python
#Calculating the percentage of students who passed math and reading (% Overall Passing)
passed_both = school_df[(school_df['reading_score'] >= 70) & (school_df['math_score'] >= 70)].groupby('school_name')['Student ID'].count()
pass_overall= round((passed_both/total_students)*100,2)
pass_overall
```

```python
raw_summary= pd.DataFrame({'School Type': school_type,
                           'Total Students':  total_students,
                           'Total School Budget': total_budget,
                           'Per Student Budget':student_budget,
                           'Average Math Score': avg_mathscore,
                           'Average Reading Score': avg_readscore,
                           'Passed Math':pass_math,
                           'Passed Reading':pass_reading,
                           'Passed Overall':pass_overall})

#formatting raw school summary
school_summary = raw_summary.copy()
school_summary['Per Student Budget'] =school_summary['Per Student Budget'].map('${:,.2f}'.format)
school_summary['Total School Budget'] =school_summary['Total School Budget'].map('${:,.2f}'.format)

school_summary
```

# Top Performing Schools (By % Overall Passing)

```python
top_performer = school_summary.nlargest(5,"Passed Overall")
top_performer
```

# Bottom Performing Schools (By % Overall Passing)

```python
bottom_performer = school_summary.nsmallest(5,"Passed Overall")
bottom_performer
```

# Math and Reading Scores by Grade

```python
# Calculating the average math score for students of 9th-12th grade at each school
math_9th = round(school_df[school_df['grade']=='9th'].groupby('school_name')['math_score'].mean(),2)
math_10th = round(school_df[school_df['grade']=='10th'].groupby('school_name')['math_score'].mean(),2)
math_11th = round(school_df[school_df['grade']=='11th'].groupby('school_name')['math_score'].mean(),2)
math_12th = round(school_df[school_df['grade']=='12th'].groupby('school_name')['math_score'].mean(),2)

# Creating a new table for the average math score by grade
math_bygrade = pd.DataFrame({'9th':math_9th,'10th':math_10th,'11th':math_11th,'12th':math_12th })

math_bygrade
```

```python
# Calculating the average reading score for students of 9th-12th grade at each school
read_9th = school_df[school_df['grade']=='9th'].groupby('school_name')['reading_score'].mean()
read_10th = school_df[school_df['grade']=='10th'].groupby('school_name')['reading_score'].mean()
read_11th = school_df[school_df['grade']=='11th'].groupby('school_name')['reading_score'].mean()
read_12th = school_df[school_df['grade']=='12th'].groupby('school_name')['reading_score'].mean()

# Creating a new table for the average reading score by grade
read_bygrade = pd.DataFrame({'9th':read_9th,'10th':read_10th,'11th':read_11th,'12th':read_12th })

# Formatting raw average reading score
read_bygrade['9th'] = read_bygrade['9th'].map('{:,.2f}'.format)
read_bygrade['10th'] = read_bygrade['10th'].map('{:,.2f}'.format)
read_bygrade['11th'] = read_bygrade['11th'].map('{:,.2f}'.format)
read_bygrade['12th'] = read_bygrade['12th'].map('{:,.2f}'.format)

read_bygrade
```

# Scores by School Spending

```python
# Identifying min and max student budget to help create bins 
student_budget.min(), student_budget.max()
```

```python
# Creating the bins in which data will be held   
bins = [0,580, 610, 640, 670]

# Creating names for the bins
group_names = ["<$580", "$580-609.9", "$610-639.9", "$640-669.9"]

school_summary["Spending Ranges (Per Student)"] = pd.cut(student_budget, bins, labels = group_names)
```

```python
# calculating avg math score, avg reading score, % passing math, % passing reading, and overall passing rate by spend
spend_group = school_summary.groupby(['Spending Ranges (Per Student)'])

math_by_spend = round(spend_group.mean()["Average Math Score"],2)
read_by_spend = round(spend_group.mean()["Average Reading Score"],2)
pass_math_spend = round(spend_group.mean()["Passed Math"],2)
pass_read_spend = round(spend_group.mean()["Passed Reading"],2)
pass_overall_spend = round(spend_group.mean()["Passed Overall"],2)
```

```python
scores_byspend = pd.DataFrame({
    "Average Math Score": math_by_spend,
    "Average Reading Score":read_by_spend,
     "% Passed Math": pass_math_spend,
    "% Passed Reading": pass_read_spend,
    "% Passed Overall": pass_overall_spend
})

scores_byspend
```

# Scores by School Size

```python
# Creating the bins in which data will be held 
size_bins = [0, 1000, 2000, 5000]

# Creating names for the bins
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_summary["School Size"] = pd.cut(school_summary["Total Students"], size_bins, labels=group_names)
```

```python
# calculating avg math score, avg reading score, % passing math, % passing reading, and overall passing rate by size
size_group = school_summary.groupby(["School Size"])

math_by_size = round(size_group.mean()["Average Math Score"],2)
reading_by_size = round(size_group.mean()["Average Reading Score"],2)
pass_math_size = round(size_group.mean()["Passed Math"],2)
pass_reading_size = round(size_group.mean()["Passed Reading"],2)
pass_overall_size = round(size_group.mean()["Passed Overall"],2)


scores_bysize = pd.DataFrame({"Average Math Score" : math_by_size,
                             "Average Reading Score": reading_by_size,
                             "% Passed Math": pass_math_size,
                              "% Passed Reading": pass_reading_size,
                              "% Passed Overall":pass_overall_size})

scores_bysize
```

# Scores by School Type

```python
#calculating avg math score, avg reading score, % passing math, % passing reading, and overall passing rate by school type
scores_bytype = school_summary[['School Type',
                                'Average Math Score',
                                'Average Reading Score',
                                'Passed Math',
                                'Passed Reading',
                               'Passed Overall']]
# Grouping by the school type
scores_bytype = round(scores_bytype.groupby('School Type').mean(),2)
scores_bytype
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
report = pd.ExcelWriter('report.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet in the report workbook.
district_summary.to_excel(report, sheet_name='District Summary')
school_summary.to_excel(report, sheet_name='School Summary')
top_performer.to_excel(report, sheet_name='top_performer')
math_bygrade.to_excel(report, sheet_name='Math Scores by Grade')
read_bygrade.to_excel(report, sheet_name='Reading Scores by Grade')
scores_byspend.to_excel(report, sheet_name='Scores by Student Budget')
scores_bysize.to_excel(report, sheet_name='Scores by Size')
scores_bytype.to_excel(report, sheet_name='Scores by Type')

report.save()
```
